```python
import pandas as pd

# Create a dataset
# Let's build the values lists
value_n_list = ['Andrew', 'Phillip', 'John', 'Laura', 'Sophia', 'Michael', 'Emma', 'Daniel', 'Olivia', 'James', 'Liam', 'Isabella', 'Lucas']
value_w_list = [50, 78, 91.5, 60, 55.2, 85.4, 48.3, 74.1, 62.0, 80.5, 77.3, 50.0, 70.2]  # [kg]
value_h_list = [1.76, 1.87, 1.69, 1.65, 1.70, 1.80, 1.58, 1.75, 1.68, 1.85, 1.79, 1.60, 1.77]  # [m]
# value_bmi_list will be calculated: bmi = w/(h^2)
# value_bmi_list = [25]*13 # init values [kg/m2]

# Let's use a dictionary ({key:value})
# with the format {"str":list}
data_dict = {"Nombre": value_n_list,
             "Peso": value_w_list,
             "Altura": value_h_list  #,
             # "BMI": value_bmi_list
             }

print(data_dict)

# Turn a dictionary into a dataframe
df = pd.DataFrame(data_dict)
print(df)
df
# promt con DataFrame df: aadd another colum
df["bmi"] = df["Peso"] / (df["Altura"] ** 2)
print(df)
df
def calssifity_bmi(bmi):
    if bmi < 18.5:
        return "Peso insuficiente"
    elif 18.5 <= bmi < 24.9:
        return "peso normal"
    elif 25 <= bmi < 29.9:
        return "Sobrepeso"
    else:
        return "Obesidad"
df["bmi"] = df["bmi"].apply(calssifity_bmi)
print(df)
df
edad = [25, 30, 22, 28, 35, 40, 27, 33, 29, 31, 26, 24, 32]
df["Edad"] = edad
print(df)
df
```
